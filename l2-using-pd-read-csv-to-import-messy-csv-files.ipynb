1. What is CSV File?
Add data files from kaggle <br>
CSV Files - A comma-separated values file is a delimited text file that uses a comma to separate values. Each line of the file is a data record. Each record consists of one or more fields, separated by commas. The use of the comma as a field separator is the source of the name for this file format.

2. Why to clean CSV files?
To get appropriate results, the data we are using has to be accurate. If, the data is not accurate it might result in wrong finding. So it is suggested to clean the data using different methods.

# 1st Approach 
* Setting header as index 1
* Dropping the last row

```python
#resetting the header of the data to index 1
df=pd.read_csv("../input/meesy-csv/EQUITY_L.csv",header=1)
```

```python
#As I can see the last line is also not corrupted, So we are dropping the last line of the data
df.drop(index=1665,axis=0)
```
